```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from keras.layers import Dense, Flatten
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
X_train[1]
```

```python
y_train
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

# **Building ANN**

```python
model = Sequential()


#Using 'Flatten layer' to convert higher dimensional array (here 28*28) into 1-D array bcz input to our neural netwrok is 784 and we have input as 28-D array so we put
#all the values side by side thus making 1-D array
model.add(Flatten(input_shape=(28,28)))

model.add(Dense(128,activation='relu'))   #here we are not giving input bcz it will automatically detect as output of Flatten will be given to 1st hidden layer
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='Adam')
```

```python
history = model.fit(X_train, y_train, epochs=25, validation_split=0.2)
```

```python
y_prob = model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
